```python
# Connexion à la base de données
import psycopg2

DB_NAME = "ecommerce"
DB_USER = "postgres"
DB_PASS = "password"
DB_HOST = "localhost"
DB_PORT = "5432"

try:
    conn = psycopg2.connect(
        database=DB_NAME,
        user=DB_USER,
        password=DB_PASS,
        host=DB_HOST,
        port=DB_PORT,
    )
    
    cur = conn.cursor()
    
    print("Connexion à la base de données réussie")
except Exception as e:
    print("Echec de connexion à la base de données :", e)
```

### **Création des tables**

```python
# Suppression des tables (si nécessaire)
delete_clients_table_query = "DROP TABLE IF EXISTS Clients"
delete_ventes_table_query = "DROP TABLE IF EXISTS Ventes"
delete_interactions_table_query = "DROP TABLE IF EXISTS Interactions"
delete_visites_table_query = "DROP TABLE IF EXISTS Visites"

try:
    cur.execute(delete_ventes_table_query)
    cur.execute(delete_interactions_table_query)
    cur.execute(delete_visites_table_query)
    cur.execute(delete_clients_table_query)
    
    conn.commit()
    
    print("Tables supprimées avec succès")

except Exception as e:
    conn.rollback()
    print("Echec lors de la suppression des tables :", e)
```

```python
# Table Clients
create_clients_table_query = """
CREATE TABLE IF NOT EXISTS Clients (
    client_id SERIAL PRIMARY KEY,
    nom TEXT,
    prenom TEXT,
    email TEXT,
    telephone TEXT,
    date_naissance DATE,
    date_creation DATE,
    score_appetence_sms REAL,
    score_appetence_email REAL,
    score_appetence_push REAL
);
"""
```

```python
# Table Ventes
create_ventes_table_query = """
CREATE TABLE IF NOT EXISTS Ventes (
    vente_id SERIAL PRIMARY KEY,
    client_id INTEGER,
    date_vente DATE,
    no_commande INTEGER,
    produit TEXT,
    quantite INTEGER,
    prix_unitaire REAL,
    sous_total REAL,
    FOREIGN KEY (client_id) REFERENCES Clients(client_id) ON DELETE CASCADE
);
"""
```

```python
# Table Interactions
create_interactions_table_query = """
CREATE TABLE IF NOT EXISTS Interactions (
    interaction_id SERIAL PRIMARY KEY,
    client_id INTEGER,
    canal TEXT,
    date_interaction DATE,
    contenu TEXT,
    resultat TEXT,
    FOREIGN KEY (client_id) REFERENCES Clients(client_id) ON DELETE CASCADE
);
"""
```

```python
# Table Visites
create_visites_table_query = """
CREATE TABLE IF NOT EXISTS Visites (
    visite_id SERIAL PRIMARY KEY,
    client_id INTEGER,
    date_visite DATE,
    page_visitee TEXT,
    duree_secondes INTEGER,
    FOREIGN KEY (client_id) REFERENCES Clients(client_id) ON DELETE CASCADE
);
"""
```

```python
# Création des tables
try:
    cur.execute(create_clients_table_query)
    cur.execute(create_ventes_table_query)
    cur.execute(create_interactions_table_query)
    cur.execute(create_visites_table_query)

    conn.commit()
    
    print("Tables créées avec succès")

except Exception as e:
    conn.rollback()
    print("Echec lors de la création des tables :", e)
```
